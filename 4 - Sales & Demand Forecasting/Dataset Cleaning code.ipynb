Converting the columns names into correct order

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

Converting Order Date and Ship Date to datetime format

```python
df["order_date"] = pd.to_datetime(df['order_date'],errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'],errors='coerce')
```

```python
df
```

 Verifying Sales, Profit, and Quantity are numeric

```python
df['sales'] = pd.to_numeric(df['sales'],errors = 'coerce')
df['profit'] = pd.to_numeric(df['profit'],errors = 'coerce')
df['quantity'] = pd.to_numeric(df['quantity'],errors = 'coerce')
```

Creating additional date-related columns

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['Month Name'] = df['order_date'].dt.month_name()
df['day'] = df['order_date'].dt.day
df['Quarter'] = df['order_date'].dt.quarter
```
