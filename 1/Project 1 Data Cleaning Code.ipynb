```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
df.rename(
    columns={'product_sub-category':'product_sub_category'},
    inplace=True
)
```

```python
df["product_base_margin"] = df["product_base_margin"].fillna(
    df["product_base_margin"].median()
)
df.head()
```

```python
numeric_cols = [
    "order_quantity",
    "sales",
    "profit",
    "discount",
    "unit_price",
    "shipping_cost",
    "product_base_margin"
]
df[numeric_cols].head()
```

```python
df['discount'] = df['discount']*100
df['product_base_margin'] = df['product_base_margin']*100
df[numeric_cols].head()
```

```python
df['sales'] = df['sales'].astype(int)
df['profit'] = df['profit'].astype(int)
df['discount'] = df['discount'].astype(int)
df['unit_price'] = df['unit_price'].astype(int)
df['shipping_cost'] = df['shipping_cost'].astype(int)
df['product_base_margin'] = df['product_base_margin'].astype(int)
df.head()
```

```python
df['order_date'].unique()
```

```python
df['order_date'].astype(str)
```

```python
def fix_date(date_str):
    date_str = str(date_str).replace("/", "-")

    parts = date_str.split("-")
    if len(parts) != 3:
        return date_str

    p1, p2, p3 = parts

    try:
        p1 = int(p1)
        p2 = int(p2)
    except:
        return date_str

    # If first part is month and second part is day
    if p1 <= 12 and p2 > 12:
        return f"{p2:02d}-{p1:02d}-{p3}"

    # Otherwise leave unchanged
    return date_str

df['order_date'] = df['order_date'].apply(fix_date)
df['ship_date'] = df['ship_date'].apply(fix_date)
```

```python
df['order_date'].head(10)
print(df['ship_date'].head(10))
```

```python
df['order_date'] = pd.to_datetime(
    df['order_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

df['ship_date'] = pd.to_datetime(
    df['ship_date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)
```
